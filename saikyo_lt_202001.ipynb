```python
base_df = pd.read_csv('2019_delay_information.csv')
```

```python
base_df.head()
```

```python
base_df.info()
```

```python
nov_df = base_df[base_df['日付'].str.contains('11月')].sort_values('日付')
nov_df
```

```python
dec_df = base_df[base_df['日付'].str.contains('12月')].sort_values('日付')
dec_df
```

```python
nov_df.describe()
```

```python
dec_df.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 11月の遅延時間
nov = nov_df['7時～10時']

# 12月の遅延時間
dec = dec_df['7時～10時']

# 11月、12月の遅延時間のタプル
delay = (nov, dec)

# 箱ひげ図
fig, ax = plt.subplots()

bp = ax.boxplot(delay)
ax.set_xticklabels(['nov', 'dec'])

plt.title('Box plot')
plt.xlabel('month')
plt.ylabel('delay time')
# Y軸のメモリのrange
plt.ylim([0,70])
plt.grid()

# 描画
plt.show()
```

```python
# Nov Data
df=pd.DataFrame({'x': nov_df['日付'], 'y1': nov_df['7時～10時'] })
 
# multiple line plot
fig = plt.figure(figsize=(6.0, 3.0))
ax = fig.add_subplot(111)
ax.set_position([0.5,0.1,4,0.8])

plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.legend()
```

```python
# Dec Data
df=pd.DataFrame({'x': dec_df['日付'], 'y1': dec_df['7時～10時'] })
 
# multiple line plot
fig = plt.figure(figsize=(6.0, 3.0))
ax = fig.add_subplot(111)
ax.set_position([0.5,0.1,4,0.8])

plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.legend()
```
