```python
import pandas as pd
```

```python
base_df = pd.read_csv('2019_delay_information.csv')
```

```python
base_df.head()
```

```python
base_df.info()
```

```python
nov_df = base_df[base_df['日付'].str.contains('11月')].sort_values('日付')
nov_df
```

```python
dec_df = base_df[base_df['日付'].str.contains('12月')].sort_values('日付')
dec_df
```

```python
nov_df.describe()
```

```python
dec_df.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 数学の点数
nov = nov_df['7時～10時']

# 国語の点数
dec = dec_df['7時～10時']

# 点数のタプル
delay = (nov, dec)

# 箱ひげ図
fig, ax = plt.subplots()

bp = ax.boxplot(delay)
ax.set_xticklabels(['nov', 'dec'])

plt.title('Box plot')
plt.xlabel('month')
plt.ylabel('delay time')
# Y軸のメモリのrange
plt.ylim([0,70])
plt.grid()

# 描画
plt.show()
```
